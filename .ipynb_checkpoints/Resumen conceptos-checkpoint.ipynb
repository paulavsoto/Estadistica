```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generar datos de una muestra
np.random.seed(42)
muestra = np.random.normal(loc=5, scale=2, size=100)

# Estimador de la media poblacional
estimador_media = np.mean(muestra)

# Intervalo de confianza para la media con un nivel del 95%
intervalo_confianza = stats.norm.interval(0.95, loc=np.mean(muestra), scale=stats.sem(muestra))

# Distribución de la variable (histograma)
plt.hist(muestra, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.title('Distribución de la Muestra')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()

# Estadístico de prueba (t-test para la media)
t_statistic, p_value = stats.ttest_1samp(muestra, 5)
print(f'Estadístico de prueba: {t_statistic}, Valor p: {p_value}')

# Prueba de hipótesis
nivel_significancia = 0.05
if p_value < nivel_significancia:
    print('Se rechaza la hipótesis nula')
else:
    print('No hay evidencia para rechazar la hipótesis nula')

# Graficar el intervalo de confianza
plt.hist(muestra, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(estimador_media, color='red', linestyle='dashed', linewidth=2, label='Media Estimada')
plt.axvline(intervalo_confianza[0], color='green', linestyle='dashed', linewidth=2, label='Intervalo de Confianza')
plt.axvline(intervalo_confianza[1], color='green', linestyle='dashed', linewidth=2)
plt.title('Distribución de la Muestra con Intervalo de Confianza')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, ttest_1samp

# Generar datos de muestra
np.random.seed(42)
datos_muestra = np.random.normal(loc=50, scale=10, size=30)

# 1. Estimador (media muestral)
estimador_media = np.mean(datos_muestra)

# 2. Intervalo de confianza (intervalo de confianza del 95%)
error_estandar = np.std(datos_muestra, ddof=1) / np.sqrt(len(datos_muestra))
intervalo_confianza = t.interval(0.95, len(datos_muestra)-1, loc=np.mean(datos_muestra), scale=error_estandar)

# 3. Distribución de la variable (supongamos una distribución normal)
distribucion_normal = np.random.normal(loc=50, scale=10, size=10000)

# 4. Estadístico de prueba (t-statistic)
t_statistic, p_valor = ttest_1samp(datos_muestra, 50)

# 5. Prueba de hipótesis (ejemplo: prueba de hipótesis nula de que la media es 50)
nivel_significancia = 0.05
if p_valor < nivel_significancia:
    resultado_prueba = "Rechazar la hipótesis nula: la media no es 50."
else:
    resultado_prueba = "No hay suficiente evidencia para rechazar la hipótesis nula."

# Gráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Histograma de la muestra
axs[0, 0].hist(datos_muestra, bins=10, color='blue', alpha=0.7)
axs[0, 0].set_title('Histograma de la Muestra')
axs[0, 0].axvline(x=estimador_media, color='red', linestyle='--', label=f'Media Muestral = {estimador_media:.2f}')
axs[0, 0].legend()

# Distribución teórica (normal)
axs[0, 1].hist(distribucion_normal, bins=50, color='green', alpha=0.7)
axs[0, 1].set_title('Distribución Teórica (Normal)')
axs[0, 1].axvline(x=50, color='red', linestyle='--', label='Media Teórica = 50')
axs[0, 1].legend()

# Intervalo de confianza
axs[1, 0].hist(datos_muestra, bins=10, color='blue', alpha=0.7)
axs[1, 0].set_title('Intervalo de Confianza (95%)')
axs[1, 0].axvline(x=intervalo_confianza[0], color='green', linestyle='--', label='Límite Inferior')
axs[1, 0].axvline(x=intervalo_confianza[1], color='orange', linestyle='--', label='Límite Superior')
axs[1, 0].legend()

# Resultado de la prueba de hipótesis
axs[1, 1].axis('off')
axs[1, 1].text(0.1, 0.5, resultado_prueba, fontsize=12)

plt.tight_layout()
plt.show()
```
