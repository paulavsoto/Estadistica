```python
import seaborn as sns
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
```

## Prueba de Hipotesis para la media poblacional 

### Con varianza desconocida

```python
# Datos de la muestra
data = np.array([7.1, 4.2, 6.4, 5.5, 7, 3.1, 8.4])

# Parámetros de la prueba de hipótesis
h0_mean = 4.5  # Hipótesis nula: media igual a 4.5
alpha = 0.05   # Nivel de significancia al 5%


# Calcular el promedio (media)
promedio = np.mean(data)

# Calcular la varianza
varianza = np.var(data)

# Calcular la desviación estándar
desviacion_estandar = np.std(data)

# Obtener el número de muestras
num_muestra = len(data)

# Imprimir resultados
print("Promedio:", promedio)
print("Varianza:", varianza)
print("Desviación Estándar:", desviacion_estandar)
print("Número de Muestra:", num_muestra)

# Realizar la prueba de hipótesis
t_statistic, p_value = stats.ttest_1samp(data, h0_mean)
print("El valor t_statistic es el valor T", t_statistic)

# Calcular los grados de libertad
df = len(data) - 1
print ("Los grados de libertad son: ", df)
# Calcular el valor crítico
t_critical = stats.t.ppf(1 - alpha, df)
print("El valor t_critical es el valor obtenido en la tabla, segun grados de libertad y significancia", t_critical)


# Imprimir resultados
print(f"Valor T: {t_statistic}")
print(f"Valor crítico: {t_critical}")
print(f"P-valor: {p_value}")

# Realizar el gráfico
x = np.linspace(-3, 3, 1000)
y = stats.t.pdf(x, df)

plt.plot(x, y, label='Distribución t')
plt.axvline(t_statistic, color='red', linestyle='--', label='Valor T')
plt.axvline(t_critical, color='blue', linestyle='--', label='Valor Crítico')
plt.fill_betweenx(y, x, t_critical, where=(x > t_critical), color='blue', alpha=0.3, label='Zona Crítica')

plt.title('Prueba de Hipótesis - Media')
plt.xlabel('Valor T')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.show()

# Imprimir el valor T y la decisión de la prueba
print(f"Valor T: {t_statistic}")
if t_statistic > t_critical: #coincide con la formula del video de coursera
    print("Rechazar la hipótesis nula")
else:
    print("No se puede rechazar la hipótesis nula")
```

### Prueba de hipotesis para la proporción o probabilidad

```python
from scipy.stats import bernoulli
import numpy as np
from scipy.stats import bernoulli, norm


# Definir parámetros
n = 50  # Tamaño de la muestra
x = 10  # Número de éxitos
alpha = 0.05  # Nivel de significancia
p_null = 0.15  # Hipótesis nula
alternative = 'two-sided'  # Hipótesis alternativa (en este caso, p es distinto de 0.15)

# Paso 1: Calcular el estadístico de prueba
p_hat = x / n
z_stat = (p_hat - p_null) / np.sqrt((p_null * (1 - p_null)) / n)

# Paso 2: Calcular el valor crítico
# Calcular los valores críticos para ambas colas
z_critical_right = norm.ppf(1 - alpha / 2)
z_critical_left = -norm.ppf(1 - alpha / 2)


# Paso 3: Realizar la prueba de hipótesis
p_value = 1 * (1 - bernoulli.cdf(np.abs(z_stat), p_null))

# Imprimir resultados
print(f"Estadístico de prueba: {z_stat}")
print(f"Valor crítico derecho: {z_critical_right}")
print(f"Valor crítico izquierdo: {z_critical_left}")
print(f"P-valor: {p_value}")

# Comparar con el nivel de significancia
if z_stat < z_critical_left or z_stat  > z_critical_right:
    print("Se rechaza la hipótesis nula.")
else:
    print("No se rechaza la hipótesis nula.")

# Crear el gráfico
x_values = np.linspace(-3, 3, 1000)
y_values = norm.pdf(x_values, 0, 1)

plt.plot(x_values, y_values, label='Distribución Normal Estándar')
plt.axvline(z_stat, color='red', linestyle='--', label='Estadístico de prueba')
plt.axvline(z_critical_right, color='green', linestyle='--', label='Valor Crítico Derecho')
plt.axvline(z_critical_left, color='blue', linestyle='--', label='Valor Crítico Izquierdo')

plt.fill_between(x_values, 0, y_values, where=(x_values > z_critical_right) | (x_values < z_critical_left), color='gray', alpha=0.5, label='Zona crítica')

plt.legend()
plt.xlabel('Z')
plt.ylabel('Densidad de probabilidad')
plt.title('Prueba de hipótesis para la probabilidad en una distribución de Bernoulli')
plt.show()
```

## Prueba de hipótesis para la varianza

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Datos de la muestra
data = np.array([2163, 2250, 2190, 2169, 2160, 2160, 2194, 2161, 2125])

# Definir parámetros
alpha = 0.05  # Nivel de significancia
df = len(data) - 1  # Grados de libertad

# Paso 1: Calcular el estadístico de prueba
sample_var = np.var(data, ddof=1)
h0_var = 500
chi_stat = (df * sample_var) / h0_var

print(sample_var)
print(df)

print(f'{df} * {sample_var} / {h0_var}')

# Paso 2: Calcular el valor crítico
chi_critical = chi2.ppf(1 - alpha, df)

# Paso 3: Realizar la prueba de hipótesis
p_value = 1 - chi2.cdf(chi_stat, df)

# Imprimir resultados
print(f'Estadístico de prueba: {chi_stat:.4f}')
print(f'Valor crítico: {chi_critical:.4f}')
print(f'P-valor: {p_value:.4f}')

# Comparar con el nivel de significancia
if chi_stat > chi_critical:
    print('Se rechaza la hipótesis nula.')
else:
    print('No se rechaza la hipótesis nula.')

# Crear el gráfico
x_values = np.linspace(0, 30, 1000)
y_values = chi2.pdf(x_values, df)

plt.plot(x_values, y_values, label='Distribución Chi-cuadrado')
plt.axvline(chi_stat, color='red', linestyle='--', label='Estadístico de prueba')
plt.axvline(chi_critical, color='green', linestyle='--', label='Valor Crítico')
plt.fill_between(x_values, 0, y_values, where=(x_values > chi_critical), color='gray', alpha=0.5, label='Zona crítica')

plt.legend()
plt.xlabel('Chi-cuadrado')
plt.ylabel('Densidad de probabilidad')
plt.title('Prueba de hipótesis para la varianza')
plt.show()
```
